# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get the ConvertedComp data
QUERY = "SELECT ConvertedComp FROM master"
df = pd.read_sql_query(QUERY, conn)

# Step 3: Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Close the database connection
conn.close()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get the Age data
QUERY = "SELECT Age FROM master"
df = pd.read_sql_query(QUERY, conn)

# Step 3: Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'].dropna(), vert=False)  # Horizontal box plot
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(axis='x', alpha=0.75)
plt.show()

# Close the database connection
conn.close()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get Age, WorkWeekHrs, and CodeRevHrs data
QUERY = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master"
df = pd.read_sql_query(QUERY, conn)

# Step 3: Drop rows with missing values in Age, WorkWeekHrs, or CodeRevHrs
df = df.dropna(subset=['Age', 'WorkWeekHrs', 'CodeRevHrs'])

# Step 4: Create a scatter plot of Age vs WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='orange')
plt.title('Scatter Plot of Age vs. Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid()
plt.show()

# Step 5: Create a bubble plot of WorkWeekHrs vs CodeRevHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], 
            s=df['Age'] * 10,  # Bubble size based on Age
            alpha=0.5, color='blue', edgecolor='black')
plt.title('Bubble Plot of Work Week Hours vs. Code Review Hours')  # Fixed title string
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid()
plt.show()

# Close the database connection
conn.close()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get the databases respondents wish to learn next year
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
top_databases = pd.read_sql_query(QUERY, conn)

# Step 3: Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases['Count'], 
        labels=top_databases['DatabaseDesireNextYear'], 
        autopct='%1.1f%%', 
        startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()

# Close the database connection
conn.close()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs 
FROM master 
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)

# Step 3: Calculate the median values
median_values = df[['WorkWeekHrs', 'CodeRevHrs']].median()

# Step 4: Create the stacked chart
plt.figure(figsize=(10, 6))
plt.barh(['Median WorkWeekHrs', 'Median CodeRevHrs'], 
         [median_values['WorkWeekHrs'], median_values['CodeRevHrs']], 
         color=['skyblue', 'lightgreen'])

plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs (Ages 30 to 35)')
plt.xlabel('Median Hours')
plt.grid(axis='x', alpha=0.75)
plt.show()

# Close the database connection
conn.close()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get ConvertedComp and Age for ages 45 to 60
QUERY = """
SELECT Age, ConvertedComp 
FROM master 
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)

# Step 3: Calculate the median ConvertedComp for each age
median_comp_by_age = df.groupby('Age')['ConvertedComp'].median().reset_index()

# Step 4: Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedComp'], marker='o', color='blue')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid()
plt.xticks(range(45, 61))  # Set x-ticks to show each age from 45 to 60
plt.show()

# Close the database connection
conn.close()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes her

# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query to get the MainBranch data
QUERY = "SELECT MainBranch FROM master"
df = pd.read_sql_query(QUERY, conn)

# Step 3: Count occurrences of each Main Branch
main_branch_counts = df['MainBranch'].value_counts()

# Step 4: Create the horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Distribution of Main Branches')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(axis='x', alpha=0.75)
plt.show()

# Close the database connection
conn.close()
```
